## Punto 0

```python
import pandas as pd 
```

```python
#datos de RRSS 2020
fbk = ['Facebook', 2449, True, 2006]
twt = ['Twitter', 339, False, 2006]
ig = ['Instragram', 1000, True, 2010]
yt = ['YouTube', 2000, False, 2005]
lkn = ['LinkedIn', 663, False, 2003]
wsp = ['WhatsApp', 1600, True, 2009]

#lista con datos
lista_rrss = [fbk, twt, ig, yt, lkn, wsp]

#crear dataframe a partir de listas
df_rrss = pd.DataFrame(lista_rrss, columns = ['Nombre', 'Cantidad', 'ES_FBK', 'Año'])
df_rrss
```

```python
# Matplotlib  basic 101
```

```python
#importar matplotlib
import matplotlib.pyplot as plt
```

```python
#ejemplo básico
x = [1,2,3,4,5]
y = [1,8,27,64,125]

plt.plot(x,y)
plt.show()
```

## Punto 1

```python
# gráfico de líneas utilizando matplotlib
plt.plot(df_rrss['Nombre'], df_rrss['Cantidad'])
plt.show()
```

```python
# gráfico de puntos utilizando matplolib
plt.scatter(df_rrss['Nombre'], df_rrss['Cantidad'])
plt.show()
```

```python
#EXTRA: como hacer lo mismo con métodos de un dataframe de pandas
df_rrss.plot(kind='scatter', x= 'Nombre', y= 'Cantidad')
plt.show()
```

## Punto 2

Hacer gráficas de barras

```python
# gráfico de barras utilizando matplotlib
plt.bar(df_rrss['Nombre'], df_rrss['Cantidad'])
plt.show()
```

```python
# gráfico de barras ordenado
df_rrss_sort = df_rrss.sort_values('Cantidad', ascending=False)
plt.bar(df_rrss_sort['Nombre'], df_rrss_sort['Cantidad'])
plt.show()
```

```python
# gráfico de barras ordenado y con color
plt.bar(df_rrss_sort['Nombre'], df_rrss_sort['Cantidad'],
        color = ['b', 'r', 'g', 'm', 'k', 'c'])
plt.show()
```

```python
# COLORES
# b: blue
# g: green
# r: red
# c: cyan
# m: magenta
# y: yellow
# k: black
# w: white
```

## Punto 3

Hacer gráficas de pastel/pie

```python
# gráfico de pie utilizando matplotlib
plt.pie(df_rrss['Cantidad'],
        labels = df_rrss['Nombre'])
plt.show()
```

```python
lista_col_rrss=['#3b5998', '#FF0000', '#25d366', "#d424b7", '#0e76a8', '#00acee']
```

```python
# gráfico de puntos utilizando matplotlib
plt.pie(df_rrss_sort['Cantidad'],
        labels = df_rrss_sort['Nombre'],
        colors = lista_col_rrss)
plt.show()
```
