# Linear Regression

```python
from scipy.stats import linregress
import matplotlib.pyplot as plt


def import_csv(file):
    return pd.read_csv(file)

path = "../data/Terminos_lagoon_TA_DIC_2023_RawData.csv"

data = import_csv(path)
```

```python


# --- 1. sal_psu vs temp_c ---
x1 = data['temp_c']
y1 = data['sal_psu']

res1 = linregress(x1, y1)
print('Regresión sal_psu vs temp_c:')
print(f'Pendiente: {res1.slope:.4f}')
print(f'Intercepto: {res1.intercept:.4f}')
print(f'R²: {res1.rvalue**2:.4f}')
print(f'p-valor: {res1.pvalue:.4e}')

# Graficar
plt.figure(figsize=(8,6))
plt.scatter(x1, y1, alpha=0.6, label='Data')
plt.plot(x1, res1.intercept + res1.slope*x1, color='red', label='Fitted line')
plt.title('Regresión Salinity vs Temperature')
plt.xlabel('Temperature (°C)')
plt.ylabel('Salinity (PSU)')
plt.legend()
plt.tight_layout()
plt.show()


# --- 2. dic_micromol_kg vs sal_psu ---
x2 = data['sal_psu']
y2 = data['dic_micromol_kg']

res2 = linregress(x2, y2)
print('\nRegresión dic_micromol_kg vs sal_psu:')
print(f'Pendiente: {res2.slope:.4f}')
print(f'Intercepto: {res2.intercept:.4f}')
print(f'R²: {res2.rvalue**2:.4f}')
print(f'p-valor: {res2.pvalue:.4e}')

# Graficar
plt.figure(figsize=(8,6))
plt.scatter(x2, y2, alpha=0.6, label='Data', color='orange')
plt.plot(x2, res2.intercept + res2.slope*x2, color='red', label='Fitted line')
plt.title('Regresión DIC vs Salinity')
plt.xlabel('Salinity (PSU)')
plt.ylabel('DIC (µmol/kg)')
plt.legend()
plt.tight_layout()
plt.show()
```
