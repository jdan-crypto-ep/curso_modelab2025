# Plot histogram of DIC and TA

```python
import pandas as pd
```

```python
def import_csv(file):
    return pd.read_csv(file)
```

```python
path = "../data/Terminos_lagoon_TA_DIC_2023_RawData.csv"
```

```python
data = import_csv(path)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Histograma para ta_micromol_kg
sns.histplot(data['ta_micromol_kg'], bins=30, kde=False, color='skyblue')
plt.title('Histograma de TA (µmol/kg)')
plt.xlabel('TA (µmol/kg)')
plt.ylabel('Frecuencia')
plt.show()

# Histograma para dic_micromol_kg
sns.histplot(data['dic_micromol_kg'], bins=30, kde=False, color='salmon')
plt.title('Histograma de DIC (µmol/kg)')
plt.xlabel('DIC (µmol/kg')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Gráfico de Boxplot (cajas)
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.boxplot(
    data=data,                 # DataFrame original con datos crudos
    x='season',                # eje X: estaciones
    y='dic_micromol_kg',       # eje Y: valores crudos
    hue='area',                # color por área
)
plt.title('Distribución de DIC (µmol/kg) por estación y área')
plt.xlabel('Estación')
plt.ylabel('DIC (µmol/kg)')
plt.legend(title='Área')
plt.tight_layout()
plt.show()
```
