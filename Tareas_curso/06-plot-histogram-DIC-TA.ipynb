# Plot histogram of DIC and TA

```python
import pandas as pd
```

```python
def import_csv(file):
    return pd.read_csv(file)
```

```python
path = "../data/Terminos_lagoon_TA_DIC_2023_RawData.csv"
```

```python
data = import_csv(path)
print(data.columns)
```

```python
import matplotlib.pyplot as plt

# Histograma de ta_micromol_kg
plt.hist(data['ta_micromol_kg'], bins=30, color='green', alpha=0.7, edgecolor='black')
plt.title('Histograma de TA (µmol/kg)')
plt.xlabel('TA (µmol/kg)')
plt.ylabel('Frecuencia')
plt.show()

# Histograma de dic_micromol_kg
plt.hist(data['dic_micromol_kg'], bins=30, color='orange', alpha=0.7, edgecolor='black')
plt.title('Histograma de DIC (µmol/kg)')
plt.xlabel('DIC (µmol/kg)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Histogramas en una sola grafica
plt.hist(data['ta_micromol_kg'], bins=30, alpha=0.5, label='TA')
plt.hist(data['dic_micromol_kg'], bins=30, alpha=0.5, label='DIC')
plt.title('Histograma TA vs DIC (µmol/kg)')
plt.xlabel('Concentración (µmol/kg)')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```
