Ozone ($O3$) is a triatomic allotropic species of oxygen, composed of three atoms of said element. Its environmental and health influence is dual and depends on its altitudinal distribution in the atmosphere. In the stratosphere—where it forms the "ozone layer"—it acts as a natural filter of ultraviolet (UV) radiation, attenuating the biologically active fraction of UV-B wavelengths. In contrast, in the troposphere (surface level), ozone is classified as a secondary pollutant with high reactivity and oxidative potential. Unlike primary pollutants, tropospheric ozone is not released directly into the atmosphere, but is formed in situ through photochemical processes. It is a key component of photochemical smog, a product of the interaction between precursors—mainly nitrogen oxides ($NOx$) and volatile organic compounds (VOCs)—under conditions of intense solar radiation. These precursors predominantly derive from the combustion of fossil hydrocarbons in motorized transport, thermoelectric plants, and various industrial activities

The overall reaction for the formation of tropospheric ozone can be expressed as:

$COV + NOₓ + hν (solar radiation) → O₃$

$hν$ represents the photonic energy necessary to initiate the photolysis of $NO2$​ and subsequent radical reactions that lead to the generation of ozone.

**Effects on human health**

The inhalation of ozone is highly harmful due to its powerful oxidizing action, which causes inflammation and cellular damage to the respiratory system. Its effects can manifest both in the short and long term.

* Short term: Irritation of the respiratory tract (cough, chest pain, dyspnea), reduced lung function, and exacerbation of asthma, bronchitis, or emphysema.
* Long term: Irreversible lung damage, increased risk of chronic diseases (childhood asthma, cardiovascular pathologies), and systemic inflammation with a possible link to neurodegenerative diseases.

Ozone monitoring in the Metropolitan Area of the Valley of Mexico (MAVM)
In the Metropolitan Area of the Valley of Mexico (MAVM), air quality surveillance systems have been implemented for several decades with the aim of evaluating and mitigating problems derived from atmospheric pollution. The first air quality measurements in the region began in the 1960s, with a limited scope and focused mainly on sulfur dioxide ($SO2$) and total suspended particulates (TSP). During the 1970s, a manual network of monitoring stations was established. However, the consolidation of an automatic, continuous, and more comprehensive system—capable of measuring multiple pollutants, including ozone ($O3$)—was achieved in the late 1980s, marking the beginning of the Automatic Atmospheric Monitoring Network.

Currently, the Atmospheric Monitoring System of Mexico City (SIMAT), attached to the Secretariat of the Environment of Mexico City, operates a network of 48 stations strategically distributed in the MAVM to offer a detailed spatial and temporal characterization of air quality. The determination of ozone concentration at these stations is carried out automatically, predominantly using the method of ultraviolet (UV) radiation absorption. This procedure consists of aspirating a sample of ambient air and exposing it to a UV light source; the ozone present in the sample absorbs a specific fraction of that radiation, and the magnitude of the absorption is directly proportional to the concentration of ozone in the air (Beer-Lambert Law).

**Air quality indexes and associated ozone risk**

The health risk derived from exposure to ozone in the MAVM is evaluated using the AIRE Y SALUD Index, an official instrument that classifies risk levels based on atmospheric ozone concentrations, expressed in parts per million (ppm), parts per billion (ppb), or micrograms per cubic meter ($μg/m3$). This index is regulated by the Mexican Official Standard NOM-172-SEMARNAT-2023, which establishes the criteria for communicating risk to the population. In the case of ozone, the classification includes five air quality ranges or "bands," each associated with a risk level and specific recommendations for the population. These values are determined from both the one-hour average concentration and the eight-hour mobile average concentration, in accordance with the guidelines of the World Health Organization (WHO).

The calculation of the air quality index for ozone is based on temporary averages that allow for the evaluation of both acute and chronic exposures, thus differentiating the adverse effects on health.

* Hourly average (1 hour): This indicator is essential for identifying instantaneous peaks of ozone concentration. The use of the hourly average enables an immediate response, facilitating the activation of alert and risk communication protocols to the population, as it directly reflects the current situation.
* Eight-hour mobile average: This value allows for the estimation of accumulated and prolonged exposure to ozone. The eight-hour mobile average is updated every hour, calculated from the average of the current hour and the previous seven. This methodology provides a more representative picture of the medium- and long-term health risk.

**Urban Stations:** Located in the city center, they are characterized by high population density, intense vehicular traffic, and local industries. They allow for the monitoring of direct population exposure to ozone peaks. In these areas, the concentrations of nitrogen oxides $(NOx)$ —mainly nitric oxide $(NO)$  from vehicles—are very high. This NO "consumes" nearby ozone from emission sources through the reaction $NO + O₃ → NO₂ + O₂$ (known as "ozone titration"), so the concentrations measured at these stations may be lower than in suburban or rural areas. Monitoring here is essential to assess the immediate health risk in densely populated areas.

**Suburban Stations:** Located on the outskirts of large cities, they have lower traffic and industrial density but are influenced by the transport of pollutants from the urban center. They are key to measuring maximum ozone values, as in these areas, the precursors (NOx and volatile organic compounds) have had enough time to react under solar radiation. They often record the highest concentrations of ozone, making them fundamental for evaluating the impact of urban pollution on surrounding areas and for issuing air quality alerts.

**Rural Stations:** Located in agricultural or remote areas, far from large sources of local pollution. They allow for the estimation of background or regional ozone levels, which are not influenced by direct local emissions but by long-distance transport (even at a continental level). They are indispensable for evaluating the effects of ozone on agriculture and wild flora, understanding the large-scale transport and accumulation of pollutants, and determining the effectiveness of regional emission reduction policies.

```python
# Calcular el máximo de OZONO por día across todas las estaciones
daily_max = df.groupby('FECHA')['OZONO'].max().reset_index()

# Convertir FECHA a datetime y extraer año, mes y día
daily_max['FECHA'] = pd.to_datetime(daily_max['FECHA'])
daily_max['year'] = daily_max['FECHA'].dt.year
daily_max['month'] = daily_max['FECHA'].dt.month
daily_max['day'] = daily_max['FECHA'].dt.day

# Agrupar por mes y día, calculando el promedio de los máximos diarios
pivot_data = daily_max.groupby(['month', 'day'])['OZONO'].mean().reset_index()

# Crear una tabla pivote completa (1-12 meses, 1-31 días)
pivot = pd.pivot_table(pivot_data, values='OZONO', index='day', columns='month', aggfunc='mean')

# Verificar si la tabla pivote tiene datos válidos
if pivot.empty or pivot.isna().all().all():
    raise ValueError("La tabla pivote está vacía o contiene solo NaN. Verifique que haya datos suficientes.")

# Imprimir estadísticas básicas para depuración
print(f"Minimum and maximum OZONE values:\nMin: {pivot.min().min()}, Max: {pivot.max().max()}")

# Definir categorías y colores tipo semáforo con opacidad reducida (alpha=0.7)
categories = ['Good', 'Acceptable', 'Poor', 'Very Poor', 'Extremely Poor']
colors = [
    to_rgba("#319B16", 0.8),
    to_rgba("#F8F400", 0.8),
    to_rgba("#F3A136", 0.8),
    to_rgba("#FF0000", 0.8),
    to_rgba("#540069", 0.8)  
]
cmap = ListedColormap(colors)

# Definir límites para las categorías
max_ozono = pivot.max().max() if not pivot.isna().all().all() else 175
bounds = [0, 58, 90, 135, 175, max(max_ozono, 176)]
norm = BoundaryNorm(bounds, cmap.N)

# Crear el mapa de calor anotado
fig, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(pivot, annot=True, fmt='.1f', cmap=cmap, norm=norm, ax=ax,
            linewidths=0.5, cbar_kws={'label': 'Categorías de Ozono'},
            mask=pivot.isna())  # Oculta celdas con NaN

# Personalizar la barra de colores
cbar = ax.collections[0].colorbar
centers = [29, 74, 112.5, 155, (175 + bounds[-1]) / 2]
cbar.set_ticks(centers)
cbar.set_ticklabels(categories)

# Etiquetas de ejes
ax.set_xlabel('Months')
ax.set_ylabel('Days')

# Título
plt.title('Annotated Heat Map of Daily Maximum Ozone (Average per Month/Day)')

# Mostrar el gráfico
plt.show()
```

```python
# Function to create plot for a specific date
def plot_ozone_for_date(month_day, date_label):
    # Filter data for specific date
    filtered_data = df[df['FECHA_HORA'].dt.strftime('%m-%d') == month_day]
    
    if filtered_data.empty:
        print(f"No data available for {date_label}")
        return
    
    # Group by station and find the maximum OZONO value
    max_values = filtered_data.groupby('ESTACION')['OZONO'].max().reset_index()
    max_values = max_values.sort_values('OZONO', ascending=False)
    
    # Create the plot
    plt.figure(figsize=(12, 6))
    bars = plt.bar(range(len(max_values)), max_values['OZONO'], color='lightgreen')
    
    plt.xlabel('Stations')
    plt.ylabel('Maximum Ozone Concentration (ppb)')
    plt.title(f'Maximum Ozone Concentration by Station - {date_label}')
    plt.xticks(range(len(max_values)), max_values['ESTACION'], rotation=45, ha='right')
    
    
    # Add threshold lines
    plt.axhline(y=96, color='orange', linestyle='--', linewidth=2, label='96 ppb Threshold')
    plt.axhline(y=136, color='red', linestyle='--', linewidth=2, label='136 ppb Threshold')
    plt.axhline(y=176, color='violet', linestyle='--', linewidth=2, label='176 ppb Threshold')
    
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    return max_values

# Generate plots for both dates
print("Generating plot for May 14th...")
may_14_data = plot_ozone_for_date('05-14', 'May 14th')


# Print summary data
if may_14_data is not None:
    print("\nMay 14th Data:")
    print(may_14_data)
    
```

### **8-HOUR WINDOW ANALYSIS**

**Mann-Whitney U Test: Comparing Ozone Between Stations**

```python
# Estaciones de interés
estaciones = ['AJU', 'TLI', 'IZT']

# Extraer los datos de ozono para cada estación (eliminar NaN)
datos = {}
for estacion in estaciones:
    ozono_vals = df[df['ESTACION'] == estacion]['OZONO'].dropna()
    datos[estacion] = ozono_vals

print("MANN-WHITNEY U TEST (Pairwise Comparisons)")

# Lista para guardar resultados
resultados = []

# Comparar todas las combinaciones por pares
from itertools import combinations

for est1, est2 in combinations(estaciones, 2):
    data1 = datos[est1]
    data2 = datos[est2]
    
    # Aplicar prueba Mann-Whitney U (hipótesis alternativa: 'two-sided')
    stat, p_value = mannwhitneyu(data1, data2, alternative='two-sided')
    
    # Interpretar
    alpha = 0.05
    significativo = p_value < alpha
    
    resultado = {
        'Comparación': f"{est1} vs {est2}",
        'Estadístico U': stat,
        'p-valor': p_value,
        'Diferencia significativa': significativo
    }
    resultados.append(resultado)
    
    print(f"\n🔹 {est1} vs {est2}")
    print(f"   U Statistician: {stat:.2f}")
    print(f"   p-valor: {p_value:.6f}")
    if significativo:
        print(f"   ✅ There is a significant difference in Ozone between{est1} and {est2}")
        # Opcional: indicar cuál tiene valores mayores
        med1, med2 = data1.median(), data2.median()
        if med1 > med2:
            print(f"      ➤ {est1} has higher median Ozone ({med1:.2f} vs {med2:.2f})")
        else:
            print(f"      ➤ {est2} has higher median Ozone ({med2:.2f} vs {med1:.2f})")
    else:
        print(f"   ❌ There is no significant difference in Ozone between {est1} and {est2}")
```

### **Results**

**Relationship between Ozone and Temperature**

* Urban stations (orange): They have the lowest ozone concentrations, especially at low temperatures, because NO from traffic is actively destroying the ozone.
* Rural stations (green): They are far from major urban NOx sources. Although the temperature may be lower, the ozone that forms in the city is not "destroyed" by titration. Instead, it is transported by the wind to these areas. With less NO to consume it, ozone accumulates, and higher concentrations are recorded than in the city, even at lower temperatures. The ozone measured at these stations is the regional background ozone and the transported ozone.
* Suburban stations (blue): They act as an intermediate point. They receive the transport of pollutants from the city, but have a lower degree of ozone destruction than urban stations. Often, the maximum ozone concentrations occur in these areas, as the precursor pollutants have had time to react and form ozone while being transported, without the destructive influence of NO that dominates in the city center.

**Inverse Relationship between Ozone and NOx**

The graph (NO vs. O₃) confirms that NO is a key precursor and factor in ozone dynamics, but its behavior varies depending on the type of station. The blue regression line for the rural station (AJM) shows the least inverse relationship (the slope closest to zero) with nitrogen oxides (NOx) compared to the urban (IZT) and suburban (TLI) stations.

* Urban Station (IZT, oreange): The regression line with the most negative slope corresponds to the urban station. This indicates a strong inverse relationship between ozone and NOx. As NOx levels (from traffic and industry) increase, ozone levels decrease. This is because nitric oxide (NO) actively consumes ozone (O₃) through the reaction NO + O₃ → NO₂ + O₂.
* Suburban Station (TLI, blue): This station shows an inverse relationship, but with a less pronounced slope than the urban one. The suburban station is far enough from the city center that some of the NOx has oxidized to nitrogen dioxide (NO₂), reducing ozone destruction. Additionally, the ozone formed in the city is transported by the wind to these areas, contributing to higher ozone levels.
* Rural Station (AJM, green): This regression line has the slope closest to zero, showing the weakest inverse relationship. This confirms that in rural areas, the impact of local NOx is minimal. The ozone concentrations recorded are not so much the result of local production, but of regional background ozone levels and, above all, the ozone that has been transported from urban and suburban areas. With very little local NOx, ozone destruction by titration is almost insignificant, which makes the relationship between the two pollutants less evident.

Based on the second graph (NO vs. Ozone), yes, the hypothesis is confirmed that nitric oxide (NO) is a key precursor and factor in ozone dynamics, but its behavior is different at each type of station. The rural station presents a different behavior from the other stations.

The analysis of tropospheric ozone concentrations in the Metropolitan Area of the Valley of Mexico (MAVM), based on SIMAT data and processed in Python, allowed for the characterization of the temporal, seasonal, and spatial variability of the pollutant, as well as its relationship with precursors (NOx).

### **Conclusions**

* Tropospheric ozone in the MAVM represents a secondary pollutant of high health relevance, with temporal and spatial patterns clearly influenced by solar radiation, NOx emissions, and the type of monitoring station.
* The combination of hourly analysis, eight-hour mobile averages, and correlation with precursors allows for a comprehensive evaluation of both acute and chronic ozone exposure.
* The temporal and spatial variability of O₃ reflects the interaction of urban sources, atmospheric transport, and photochemical processes.
* Suburban and rural stations are essential for identifying maximum peaks and background ozone levels, respectively, providing key information for mitigation policies and health alerts.
